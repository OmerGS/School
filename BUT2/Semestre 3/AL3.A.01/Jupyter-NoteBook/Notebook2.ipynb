# Classification Supervisée
## 1. Arbre de Decision
### Exemple sur le dataset iris de sklearn.datasets



Dans cet exemple nous allons essayer de mettre en oeuvre l'algorithme d'arbre de décision via le module ScikitLearn.

---

## Etape 1 : Importer les dépendances

Ici nous avons utilisé sklearn pour y prendre datasets, model et tree. Nous avons egalement utilisé matplotlib pour afficher l'arbre de décision.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import random
```

#### Etape 2 : Recuperer les données puis les traiter

```python
iris = load_iris() #Recuperation du datasets contenu dans sklearn.
X = iris.data      #
y = iris.target    # 

# Separer les valeurs pour entrainer et les valeurs pour la prediction, on va mettre un random dans random_state qui sera comme la seed de l'arbre de décision.
# De plus on va recuperer 80% des données pour entrainer l'algorithme et les 20% autres seront dédiés aux tests.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random.randint(1 , 999999999)) 
```

```python
clf = DecisionTreeClassifier() # Initialise le classeur pour l'arbre de décision
clf.fit(X_train, y_train) # Fitting the classifier on the training data


# Afficher l'arbre de décision
plt.figure(figsize=(15, 10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

---

## 2. RandomForest
On va voir la précision de notre algorithme en le comparant avec d'autres arbres de décision

```python

# creating a RF classifier
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators = 100)  
 
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
 
# performing predictions on the test dataset
y_pred = clf.predict(X_test)
 
# metrics are used to find accuracy or error
from sklearn import metrics  
print()
 
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))
```

# L'Arbre de Décision (Decision Tree)


```python
sizeSample = 1500 # à baisser si l'abre ne s'affiche pas
```

```python
dataRaw =fetch_openml(data_id=43695,return_X_y=True)
data = dataRaw[0][:sizeSample]
print("n_sample=",data.shape[0])
print("n_feature=",len(data.columns))
data
```

```python
X = data[["fixed_acidity","volatile_acidity","citric_acid","residual_sugar","chlorides","free_sulfur_dioxide","total_sulfur_dioxide","density","pH","sulphates","alcohol"]] #n_sample, n_feature <=> n_row, n_ columns
Y = data["quality"]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
plt.figure(figsize=(200,60),dpi=100)
tree.plot_tree(clf,proportion=True,fontsize=12)
```
