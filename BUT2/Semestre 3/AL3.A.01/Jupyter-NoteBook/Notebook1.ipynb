# Cas d'utilisation de la régression : Modèle concernant les voitures d'occasion

Dans ce premier exemple, nous allons vous présenter un modèle de prédiction basé sur l'algorithme de la *Random Forest*, introduit lors du cours n°2. Ce modèle a pour objectif de prédire le prix de vente des voitures d'occasion à partir de certaines caractéristiques spécifiques. Voici le déroulement de l'approche :

### Étapes de réalisation du modèle :
1. **Chargement des données** : Nous allons utiliser un fichier CSV contenant 302 échantillons de voitures d'occasion.
   
2. **Séparation des données** : 
   - La moitié des données servira d'ensemble d'entraînement (dataset), c'est-à-dire les voitures dont les prix sont déjà connus. 
   - L'autre moitié servira pour les prédictions, afin d'évaluer la performance du modèle.

3. **Entraînement du modèle** : Nous allons entraîner notre modèle de régression sur l'ensemble d'entraînement.

4. **Prédiction** : Le modèle effectuera ensuite des prédictions sur l'ensemble de test, c'est-à-dire sur les voitures dont nous ne connaissons pas encore le prix.

5. **Évaluation** : Nous comparerons les prix prédits par notre modèle aux vrais prix de vente pour évaluer sa performance.

6. **Prédiction personnalisée** : Enfin, nous offrirons la possibilité de fournir les caractéristiques d'une voiture spécifique (comme le kilométrage, l'année, le type de carburant, etc.), afin que le modèle puisse estimer son prix de vente.

---

### Limites du modèle :
- **Manque de données récentes** : Le modèle ne pourra pas fournir des estimations précises pour les véhicules très récents en raison du manque de données pour les modèles après 2020.
- **Kilométrage élevé** : Les prédictions peuvent être inexactes pour des véhicules n'ayant pas un kilométrage faible (au-dessous de 100 000 km), car ces cas cnotraires ne sont pas bien représentés dans notre ensemble de données.
- **Nombre de propriétaires** : Le modèle ne tient pas compte du nombre de propriétaires précédents des véhicules, car cette information est mal représentée (toujours à 0 dans les données).
  
### Forces du modèle :
- **Variété des paramètres** : Notre modèle est bien adapté à traiter une grande variété de caractéristiques, en particulier pour la transmission (manuelle ou automatique) et le type de carburant (essence ou diesel), car les données incluent de nombreuses variétés dans ces catégories.

---

En résumé, bien que le modèle ait des limites pour les véhicules récents ou avec des kilométrages extrêmes, il reste utile pour prédire les prix de vente de voitures d'occasion dans la majorité des cas, surtout pour des modèles intermédiaires ou anciens.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import datetime
```

```python
# data import
voiture_init = pd.read_csv('carData.csv') #Importation des données via le fichier voituretureData.csv
voiture = voiture_init.copy()
display(voiture.head())
```

```python
#Transformation des données qui nous interesse comme l'Annee -> Age de la voiture.

voiture['Transmission'] = voiture['Transmission'].replace(['Manual','Automatic'],[0,1])
voiture['Type_Vendeur'] = voiture['Type_Vendeur'].replace(['Dealer','Individual'],[0,1])
voiture['Age'] = datetime.date.today().year - voiture['Annee']
voiture = voiture.drop(columns=['Voiture','Annee']).copy()
voiture = pd.get_dummies(voiture)
display(voiture.head())
```

```python
X = voiture.drop(columns='Prix').values
y = voiture['Prix'].values
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=69)
```

```python
models = [LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor()] #On va tester nos données avec plusieurs modèle de regression
model_name = []
rmse = [] #Evaluation de notre modele de regression
r_square = [] 
for model in models:
    model.fit(X_train,y_train) #On va entrainer le modele
    y_pred = model.predict(X_test) #Le modele va predire via les données qu'on avait split et qu'il n'a pas utilisé pour apprendre

    model_name.append(model.__class__.__name__)
    rmse.append(mean_squared_error(y_test,y_pred,squared=False)) #Score d'evaluation de notre modele c'est à dire, la différence +/- moyenne prévue entre une valeur prédite et la valeur réelle.
    r_square.append(r2_score(y_test,y_pred))

results = pd.DataFrame({'Models':model_name,'RMSE':rmse,'R-Squared':r_square})
results
```

```python
# Création et entraînement du modèle
regressor = RandomForestRegressor(n_estimators=300, min_samples_split=2, min_samples_leaf=1, max_depth=5)
regressor.fit(X_train, y_train)

# Prédiction
y_pred = regressor.predict(X_test)

# Affichage de la régression avec les labels
sns.regplot(x=y_test, y=y_pred, ci=95)
plt.xlabel('Prix réel des voitures')  # Label de l'axe des abscisses
plt.ylabel('Prix prédit par le modèle')  # Label de l'axe des ordonnées
plt.title('Comparaison des prix réels et prédits des voitures')  # Titre du graphique
plt.show()
```

```python
Year = 2015 # Which years model
Present_Price = 15 # prix en milliers
Kms_Driven = 100000 # Kilometrage
Seller_Type = 1 # Type de vendeur - 0 si Concesionaire,  1 si Particulier.
Transmission = 0 # Transmission - 0 si Manuel,  1 si Automatique.
Owner = 5 # Nombre de proprietaire de la voiture

# Type de Carburant, 1 pour le type choisie et 0 pour le reste.
Carburant_Diesel = 1
Carburant_Petrol = 0


#Prediction du prix de vente
Age = datetime.date.today().year - Year #Recupere l'age à partir de l'année du vehicule.
y_new = [Present_Price,Kms_Driven,Seller_Type,Transmission,Owner,Age, Carburant_Diesel,Carburant_Petrol] #On va recuperer les données entrée par l'utilisateur
y_new_scaled = scaler.transform([y_new])
Selling_Price = regressor.predict(y_new_scaled)
print('Prix de vente estimée : ',Selling_Price[0].round(2)*1000, "€")
```

# Méthode des moindres carrés
On va vérifier si la pression artériel au repos augmente avec l'âge

```python
dataRaw =fetch_openml(data_id=204,return_X_y=True)
```

### Division des données
On sépare les données en 2, une partie pour entraîné le modèle et l'autre pour le testé

```python
data = dataRaw[0]
testsample=100
Xtrain = data[["age"]].iloc[testsample:].values # age
Xtest = data[["age"]].iloc[:testsample].values # age

Ytrain = data["trestbps"].iloc[testsample:].values # taux de cholesterol
Ytest = data["trestbps"].iloc[:testsample].values # taux de cholesterol
```

### Entraînement
On entraîne le modèle avec les 100 premières données et on calcule leur tendance

```python
#entraînement
reg = linear_model.LinearRegression()
reg.fit(Xtrain,Ytrain)
#prédiction
Ypred = reg.predict(Xtest)
```

```python
Xtest_flattened = Xtest.flatten()
Ypred_flattened = Ypred.flatten()

plt.scatter(Xtest_flattened, Ytest, label="True Values")
plt.plot(Xtest_flattened, Ypred_flattened, color='red', label="Predicted Values")  # Plot predictions as a line
plt.xlabel("Age")
plt.ylabel("Pression Arterielle")
plt.legend()

plt.show()
```
